```python
import pandas as pd
from collections import defaultdict
```

```python
df = pd.read_csv('/home/mw/input/Novel4641/gutenberg_fiction_metadata.csv')
df.head()
```

```python
len(df)
```

```python
sdict = defaultdict(int)
for i in range(len(df)):
    for j in eval(df['subjects'][i]):
        sdict[j]+=1
```

```python
sorted(sdict.items(), key = lambda kv: kv[1], reverse = True)[:10]
```
